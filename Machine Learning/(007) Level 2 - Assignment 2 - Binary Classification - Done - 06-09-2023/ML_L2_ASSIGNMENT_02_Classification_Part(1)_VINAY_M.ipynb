```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading CSV Dataset
df = pd.read_csv('Loan_Default.csv')
```

```python
df
```

```python
# Checking DType
df.info()
```

```python
# Dropping Non-Significant Columns
df.drop(['ID','year','rate_of_interest','Interest_rate_spread','Upfront_charges','Gender'],axis='columns', inplace=True)
```

```python
# Checking missing values
df.isna().sum()
```

```python
# percentage of missing values
missing_values = df.isna().sum()
percentage_missing = (missing_values / len(df)) * 100
round(percentage_missing, 2)
```

```python
# We can Drop or We can go for Imputation
# For Numerical - KNNImputation
# For Categorical - Mode Imputation can be performed
```

```python
# counting categorical columns
catCol = []
for col in df.select_dtypes(include=[object]).columns:
    catCol.append(col)
print(f"{len(catCol)} out of {df.shape[1]} are categorical")
catCol
```

```python
# counting Numerical columns
numricalCol = []
for col in df.select_dtypes(include=[int, float]).columns:
    numricalCol.append(col)
print(f"{len(numricalCol)} out of {df.shape[1]} are Numerical")
numricalCol
```

```python
# Writing a function to impute Numerial and Categorial
from sklearn.impute import KNNImputer

def impute_missing_data(df, k_neighbors=5):
    dfImputed = df.copy()
    for column in dfImputed.columns:
        if dfImputed[column].dtype == 'object':
            mode_value = dfImputed[column].mode().values[0]
            dfImputed[column].fillna(mode_value, inplace=True)
        else:
            imputer = KNNImputer(n_neighbors=k_neighbors)
            column_data = dfImputed[column].values.reshape(-1, 1)
            imputed_values = imputer.fit_transform(column_data)
            dfImputed[column] = imputed_values
    return dfImputed
```

```python
imputed_df = impute_missing_data(df, k_neighbors=5)
```

```python
# percentage of missing values after imputation
missing_values = imputed_df.isna().sum()
percentage_missing = (missing_values / len(df)) * 100
round(percentage_missing, 2)
```

```python
#Saving Imputed Dataframe as it takes more time when we run 
imputed_df.to_csv('imputed_loan_default_data.csv', index=True)
```
